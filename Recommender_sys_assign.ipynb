Problem Statement: Build a recommender system by using cosine simillarties score.

```python
import pandas as pd
import numpy as np 
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
import warnings
warnings.filterwarnings("ignore")
```

```python
books = pd.read_csv("G:/data sceince/Assignments/Recommendation Systems/book (1).csv")
```

```python
books.head()
```

```python
books_df = books.rename({'User.ID':'userid','Book.Title':'booktitle','Book.Rating':'bookrating'},axis = 1)
```

```python
books_df
```

```python
books1 = books_df.drop(['Unnamed: 0'], axis = 1)
```

```python
books1
```

```python
len(books1['userid'].unique())
```

```python
# extracting unique values from Book.Title column, also getting the count of unique books
```

```python
len(books1['booktitle'].unique())
```

```python
# Creating Collaborative matrix using PIVOT_Table function
```

```python
books_df1 = books1.pivot_table(index = 'userid',
                        columns = 'booktitle',
                        values = 'bookrating').reset_index(drop = True)
```

```python
books_df1
```

```python
# Mapping the unique userids with the index  
```

```python
books_df1.index = books1.userid.unique()
```

```python
books_df1
```

```python
# As we are required to calculate the similarities so replacing all the null values with 0.0
```

```python
books_df1.fillna(0, inplace = True)
```

```python
books_df1
```

```python
# Calculating similarties using cosine metrics
```

```python
user_sim = 1 - pairwise_distances(books_df1.values, metric = 'cosine')
```

```python
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)            # Converting the calculated similarities in dataframe
user_sim_df.iloc[0:5,0:5]                       # Displaying first 5 records
```

```python
user_sim_df.index = books1.userid.unique()          # Mapping the index with original userids 
user_sim_df.columns = books1.userid.unique()        # Mapping the column_names with original userids
```

```python
user_sim_df.iloc[0:5,0:5]
```

```python
# As the similarities between the same userids will be always one so as we are required to see the similarities between different users so filling the diagonal values by 0

np.fill_diagonal(user_sim,0)         
```

```python
user_sim_df.iloc[0:5,0:5]
```

```python
# Finding out which users are having maximum simlarities between them

user_sim_df.idxmax(axis = 1)
```
